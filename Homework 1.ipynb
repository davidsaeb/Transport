```python
#parts c and d
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

#governing eqs
def fun(x, y,arg):
    Da_v,Da_s=arg
    return np.vstack((y[1], Da_v * y[0] * y[2], y[3], 2 * Da_v * y[0] * y[2], y[5], -Da_v * y[0] * y[2]))

#boundary conditions
def bc(ya,yb,arg):
    Da_v,Da_s = arg
    return np.array([ya[0]-1, ya[2]-1,ya[4], yb[1]-Da_s*yb[0]*yb[2],yb[3]-2*Da_s*yb[0]*yb[2], yb[5]+Da_s*yb[0]*yb[2]])


xg = np.linspace(0, 1, 1001)
yg = np.zeros((6, xg.size))

yg[0] = yg[2] = 1  # Initial guess 
yg[4] = 0          

Da_v = Da_s = 1  #given

# Solve 
solval = solve_bvp(lambda x, y: fun(x, y, (Da_v, Da_s)), lambda ya, yb: bc(ya, yb, (Da_v, Da_s)),xg,yg)


theta_A = solval.y[0]
dthetaAdx = solval.y[1]
theta_B = solval.y[2]
dthetaBdx = solval.y[3]
theta_C = solval.y[4] 
dthetaCdx = solval.y[5]  


plt.plot(xg, theta_A, label='cA')
plt.plot(xg, theta_B, label='cB')
plt.plot(xg, theta_C, label='cC')  
plt.xlabel('Dimensionless length')
plt.ylabel('Dimensionless concentration')
plt.legend()
plt.title('Concentration profiles')
plt.grid()
plt.show()


plt.plot(xg, dthetaAdx, label='JA')
plt.plot(xg, dthetaBdx, label='JB')
plt.plot(xg, dthetaCdx, label='JC')  
plt.xlabel('Dimensionless length')
plt.ylabel('Dimensionless flux')
plt.legend()
plt.title('flux profiles')
plt.grid()
plt.show()
```
